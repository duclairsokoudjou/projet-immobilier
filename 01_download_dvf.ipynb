```python
import requests 
import os 
from tqdm import tqdm 
import time
```

```python
Departement = "75" 
url = f"https://files.data.gouv.fr/geo-dvf/latest/csv/2023/departements/{Departement}.csv" 
output_path = f"data/raw/dvf_{Departement}_2023.csv"
```

```python
os.makedirs("data/raw/dvf", exist_ok=True)

print("Téléchargement en cours...")
response = requests.get(url)
with open(output_path, "wb") as f:
    f.write(response.content)

print("Fichier téléchargé !")
```
