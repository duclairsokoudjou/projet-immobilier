#### Calcul prix au m² (avant nettoyage) 

```python
# Nettoyer et convertir valeur_fonciere
df["Valeur fonciere"] = (
    df["Valeur fonciere"]
    .str.replace(",", ".", regex=False)   # remplace la virgule par un point
    .str.replace(" ", "", regex=False)    # supprime les espaces éventuels
)
df["Valeur fonciere"] = pd.to_numeric(df["Valeur fonciere"], errors="coerce")

df["prix_m2"] = df["Valeur fonciere"] / df["Surface reelle bati"] 
```

#### Stats par type de bien 

```python
# Garder uniquement les lignes avec un prix_m2 valide et plausible
df_propre = df[
    df["prix_m2"].notna() &          # pas de NaN
    (df["prix_m2"] > 100) &          # prix minimum plausible €/m²
    (df["prix_m2"] < 50000)          # prix maximum plausible €/m²
].copy()

# Relancer le describe sur les données propres
df_propre.groupby("Type local")["prix_m2"].describe()
```
